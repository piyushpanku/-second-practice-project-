```python
# remove the Dystopiya Residual column
data=df.drop('Dystopia Residual', axis=1)
```

```python
data
```

```python
# create the target variables
target = data['Happiness Score']
```

### Create the features

```python
features = data.drop('Happiness Score', axis=1)
```

# Train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(features, target)
```

### Prediction and results Features

```python
predictions = model.predict(features)
```

### predict the happiness score considering all the other factors mentioned in the dataset

```python
print("Mean absolute error :",np.mean(np.abs(predictions - target)))
print("Root mean squared error :", np.sqrt(np.mean(predictions - target)**2))
```

### Model Saving 

```python
import pickle
filename ='World Happiness Report '
pickle.dump(model, open(filename, 'wb'))
```
